```python
support_tickets['created_at'] = pd.to_datetime(support_tickets['created_at'], errors='coerce')
support_tickets['resolved_at'] = pd.to_datetime(support_tickets['resolved_at'], errors='coerce')
```

```python
numeric_columns = ["first_response_mins", "resolution_hours", "csat_score", "sla_target_hours", "sla_met_flag"]
for col in numeric_columns:
    support_tickets[col] = pd.to_numeric(support_tickets[col], errors='coerce')
```

```python
text_columns = ["customer_city", "channel", "priority", "issue_type", "assigned_agent", "status"]

def clean_text_cols(value):
    if pd.isna(value):
        return value
    return str(value).strip().title()



for col in text_columns:
    support_tickets[col] = support_tickets[col].apply(clean_text_cols)
```

```python
support_tickets["customer_city"].unique()
```

```python
support_tickets.drop_duplicates(inplace=True)
```

```python
support_tickets.duplicated().sum()
```

```python
support_tickets['customer_city']= support_tickets['customer_city'].fillna('Unknown')
support_tickets['assigned_agent'] = support_tickets['assigned_agent'].fillna('Unassigned')
support_tickets['csat_score'] = support_tickets['csat_score'].fillna(support_tickets['csat_score'].median())
```

```python
support_tickets.loc[support_tickets["first_response_mins"] < 0, "first_response_mins"] = np.nan
support_tickets.loc[support_tickets["resolution_hours"] < 0, "resolution_hours"] = np.nan
```

```python
sla_map={
    "Low": 72,
    "Medium": 48,
    "High": 24,
    "Critical": 8
}
support_tickets['sla_target_hours'] = support_tickets['priority'].map(sla_map)
```

```python
support_tickets.head(10)
```

```python
support_tickets['sla_met_flag'] = np.where(
    (support_tickets['resolved_at'].notna()) & (support_tickets['resolution_hours'].notna()), 
    (support_tickets['resolution_hours'] <= support_tickets['sla_target_hours']).astype(int), np.nan)
```

```python
support_tickets["Month Name"] = support_tickets["created_at"].dt.month_name()
support_tickets["Day of Week"] = support_tickets["created_at"].dt.day_name()
support_tickets.head()
```
